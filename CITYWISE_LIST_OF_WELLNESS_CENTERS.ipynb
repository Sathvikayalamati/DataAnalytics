```python
import pandas as pd
```

```python
file_path = "health scheme.csv"  
df = pd.read_csv(file_path)
```

```python
df.info()
```

```python
print(df.isnull().sum())  # Check which columns have missing values
```

```python
df["wellnesscentreaddress"] = df.groupby("cityname")["wellnesscentreaddress"].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Unknown Address")
)

df
```

```python
df["wellnesscentrecontactno"] = df.groupby("cityname")["wellnesscentrecontactno"].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "9999999999")
)
df
```

```python
print(df.isnull().sum())  # Ensure no missing values remain
```

```python
# Convert contact numbers to string
df["wellnesscentrecontactno"] = df["wellnesscentrecontactno"].astype(str)

# Convert latitude and longitude to numeric
df["latitude"] = pd.to_numeric(df["latitude"], errors="coerce")
df["longitude"] = pd.to_numeric(df["longitude"], errors="coerce")
df
```

```python
df["cityname"] = df["cityname"].str.strip().str.title()
df["wellnesscentrename"] = df["wellnesscentrename"].str.strip().str.title()
df["category"] = df["category"].str.strip().str.lower()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.countplot(y=df["cityname"], order=df["cityname"].value_counts().index, palette="viridis")
plt.xlabel("Number of Wellness Centres")
plt.ylabel("City Name")
plt.title("Distribution of Wellness Centres by City")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df["category"].value_counts().plot.pie(autopct="%1.1f%%", colors=["skyblue", "lightcoral", "gold"])
plt.title("Distribution of Wellness Centres by Category")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x=df["category"], palette="viridis")
plt.xticks(rotation=45)
plt.title("Number of Wellness Centers per Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
top_cities = df["cityname"].value_counts().nlargest(10)
sns.barplot(x=top_cities.index, y=top_cities.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.xlabel("City")
plt.ylabel("Number of Wellness Centers")
plt.title("Top 10 Cities with Most Wellness Centers")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df["category"].value_counts().plot.pie(autopct="%1.1f%%", cmap="Set3", startangle=90)
plt.title("Wellness Center Distribution by Category")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["doctorcount"], bins=10, kde=True, color="skyblue")
plt.xlabel("Number of Doctors")
plt.ylabel("Frequency")
plt.title("Distribution of Doctor Counts")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
